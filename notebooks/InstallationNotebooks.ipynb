```python
import importlib
```

```python
packages = [
    "fastf1", "pandas", "numpy", "matplotlib", "plotly",
    "streamlit", "sqlalchemy", "psycopg2", "ortools", "scipy"
]

for pkg in packages:
    try:
        module = importlib.import_module(pkg)
        print(f"{pkg}: OK (version {getattr(module, '__version__', 'unknown')})")
    except ImportError:
        print(f"{pkg}: NOT INSTALLED")
```
